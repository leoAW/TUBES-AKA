**PROGRAM MEMBANDINGKAN KINERJA ALGORITMA REKURSIF DAN ITERATIF PADA PEMBELAHAN AMOEBA UNTUK BERBAGAI NILAI INPUT N**

```python
import time
import matplotlib.pyplot as plt
from prettytable import PrettyTable
```

```python
# Rekursif
def amoeba_recursive(n):
    if n <= 1:
        return 1
    return amoeba_recursive(n - 1) + amoeba_recursive(n - 2)
```

```python
# Iteratif
def amoeba_iterative(n):
    if n <= 1:
        return 1
    a, b = 1, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b
```

```python
# Grafik untuk menyimpan data
n_values = []
recursive_times = []
iterative_times = []
```

```python
# Fungsi untuk membuat dan memperbarui grafik
def update_graph():
    plt.figure(figsize=(8, 6))
    plt.plot(n_values, recursive_times, label='Recursive', marker='o', linestyle='-')
    plt.plot(n_values, iterative_times, label='Iterative', marker='o', linestyle='-')
    plt.title('Performance Comparison: Recursive vs Iterative (Amoeba Division)')
    plt.xlabel('Input (n)')
    plt.ylabel('Execution Time (seconds)')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Fungsi untuk mencetak tabel waktu eksekusi
def print_execution_table():
    table = PrettyTable()
    table.field_names = ["n", "Recursive Time (s)", "Iterative Time (s)"]
    min_len = min(len(n_values), len(recursive_times), len(iterative_times))
    for i in range(min_len):
        table.add_row([n_values[i], recursive_times[i], iterative_times[i]])
    print(table)
```

```python
# Program utama
while True:
    try:
        print("Program berjalan...")
        n = int(input("Masukkan nilai n (atau ketik -1 untuk keluar): "))
        print(f"Input yang diberikan: {n}")
        if n == -1:
            print("Program selesai. Terima kasih!")
            break
        if n < 0:
            print("Masukkan nilai n yang positif!")
            continue

        n_values.append(n)

        # Ukur waktu eksekusi algoritma rekursif
        start_time = time.time()
        amoeba_recursive(n)
        recursive_times.append(time.time() - start_time)

        # Ukur waktu eksekusi algoritma iteratif
        start_time = time.time()
        amoeba_iterative(n)
        iterative_times.append(time.time() - start_time)

        # Cetak tabel waktu eksekusi
        print_execution_table()

        # Perbarui grafik
        update_graph()

    except ValueError:
        print("Masukkan nilai n yang valid!")
    except Exception as e:
        print(f"Terjadi kesalahan: {e}")
```

Dari tabel dan grafik yang membandingkan waktu eksekusi algoritma rekursif dan iteratif pada pembelahan amoeba untuk berbagai nilai input(n) didapatkan bahwa:

1. **Algoritma Iteratif** lebih cepat dibandingkan dengan **algoritma Rekursif**, terutama saat \( n \) kecil.
2. Saat \( n \) bertambah besar, waktu eksekusi algoritma rekursif meningkat secara signifikan, yang menunjukkan **pertumbuhan eksponensial**, sedangkan algoritma iteratif memiliki **pertumbuhan linear**.
3. Pada nilai \( n = 30 \), perbedaan performa paling terlihat dan mencolok karena algoritma rekursif memakan waktu jauh lebih lama dibandingkan iteratif.

Oleh karena itu, bisa disimpulkan bahwa algoritma iteratif lebih efisien untuk input besar dalam permasalahan pembelahan amoeba.

**Percobaan perbandingan algoritma rekursif dan iteratif yang lainnya**

```python
import time

# Fungsi rekursif untuk menghitung pembelahan amoeba
def amoeba_recursive(n):
    if n <= 1:
        return 1
    return amoeba_recursive(n - 1) + amoeba_recursive(n - 2)

# Fungsi iteratif untuk menghitung pembelahan amoeba
def amoeba_iterative(n):
    if n <= 1:
        return 1
    a, b = 1, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b

# Fungsi untuk membandingkan waktu eksekusi kedua metode
def compare_performance(n):
    print(f"Calculating amoeba divisions for n = {n}")

    # Waktu untuk metode rekursif
    start_time = time.time()
    recursive_result = amoeba_recursive(n)
    recursive_time = time.time() - start_time
    print(f"Recursive result: {recursive_result}, Time taken: {recursive_time:.6f} seconds")

    # Waktu untuk metode iteratif
    start_time = time.time()
    iterative_result = amoeba_iterative(n)
    iterative_time = time.time() - start_time
    print(f"Iterative result: {iterative_result}, Time taken: {iterative_time:.6f} seconds")

    # Kesimpulan
    print("\nPerformance Comparison:")
    print(f"Recursive time: {recursive_time:.6f} seconds")
    print(f"Iterative time: {iterative_time:.6f} seconds")
    if recursive_time > iterative_time:
        print("Iterative method is faster.")
    else:
        print("Recursive method is faster.")

# Input nilai n untuk percobaan
n = 10
compare_performance(n)
```
